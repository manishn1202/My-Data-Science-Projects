Describe the data

```
df['Sentiment'].plot(kind='hist', bins=5)
```

```
sns.set(style ="darkgrid") 
ax = sns.stripplot(x = 'Competitor_Credit', y = 'Sentiment', data = df)
```

```
sns.violinplot(x="Free Equipment",data=df)
```

```
sns.catplot("Sentiment","Discount Service Call", data=df,kind="bar",height=5, aspect=2, palette="muted",legend=False)
```

```
sns.boxplot("Sentiment", data=df)
```

```
sns.pairplot(df, kind="reg",palette="m")
```

```
sns.set(style="darkgrid", color_codes=True)
sns.jointplot(x=df["Sentiment"],y=df["Saves"], kind='scatter',s=300, color='crimson', edgecolor="gold", linewidth=2)
```

```
sns.jointplot(x=df["Sentiment"],y=df["Saves"], kind='hex')
```

```
sns.jointplot(x=df["Sentiment"],y=df["Saves"], kind='kde')
```

```
sns.kdeplot(df['Saves'],shade=True,color="r")
sns.kdeplot(df['Competitor_Credit'],shade=True,color="gold")
sns.kdeplot(df['Discount Service Call'],shade=True,color="navy")
```

```
sns.regplot(x=df["Saves"],y=df["Free Equipment"], line_kws={"color":"darkgreen","alpha":0.8,"lw":2})
sns.regplot(x=df["Saves"],y=df["Discount Service Call"], line_kws={"color":"m","alpha":0.8,"lw":2})
sns.regplot(x=df["Saves"],y=df["Sentiment"], line_kws={"color":"gold","alpha":0.8,"lw":2})
```
